```python
from __future__ import annotations
from typing import Any,Type

Null=-1

class Node:
    def __init__(self,data=Null,next=Null,dnext=Null):
        self.data=data
        self.next=next
        self.dnext=dnext ##프리리스트 포인터
        
class ArrayLinkedList:
    def __init__(self,capacity:int):
        self.head=Null
        self.current=Null
        self.max=Null
        self.deleted=Null
        self.capacity=capacity
        self.n=[Node()]*self.capacity
        self.no=0
        
    def __len__(self)->int:
        return self.no
    
    def get_insert_index(self):        ##삽입할 레코드의 """인덱스"""를 구함
        if self.deleted==Null:
            if self.max+1<self.capacity:
                self.max +=1
                return self.max
            else:
                return Null           ## 초과
        else:
            rec=self.deleted
            self.deleted=self.n[rec].dnext
            return rec
        
    def delete_index(self,idx:int)->None: ##idx를 프리리스트에 등록
        if self.deleted=Null:
            self.deleted=idx
            self.n[idx].dnext=Null
        else:
            rec=self.deleted
            self.deldted=idx
            self.n[idx].dnext=rec
    
    def search(self,data:Any)->int:
        cnt=0
        ptr=self.head
        while ptr !=Null:
            if self.n[ptr].data==data:
                self.current=ptr
                return cnt
            cnt+=1
            ptr=self.n[ptr].next
        return Null                        ##못 찾았을때
     
    def __contains__(self,data:Any)->bool:
        return self.search(data)>=0
    
    def add_first(self,data:Any):
        ptr=self.head                                ##삽입 전 머리노드
        rec=self.get_insert_index()                   
        if rec!=Null:                                  ##인덱스가 존재하면
            self.head=self.current=rec
            self.n[self.head]=Node(data,ptr)
            self.no+=1
            
    def add_last(self,data:Any)->None:
        if self.head==Null:
            self.add_first(data)
        else:
            ptr=self.head
            while self.n[ptr].next !=Null:
                ptr=self.n[ptr].next
            rec=self.get_insert_index()                ## ?????
            
    def remove_first(self)->None:
        if self.head!=Null:                           ##리스트가 head값이 안비어있으면
            ptr=self.n[self.head].next                ##delete_index수행
            self.delete_index(self.head)
            self.head=self.current=ptr
            self.no-=1
            
    def remove_last(self)->None:
        if self.head!=Null:
            if self.n[self.head].next ==Null:
                self.remove_first()
            else:
                ptr=self.head                    ##스캔 중인 노드
                pre=self.head                    ## ptr 의 앞쪽 노드
                
                while self.n[ptr].next !=Null:
                    pre=ptr                                   ##ptr의 앞쪽 노드 선언
                    ptr=self.n[ptr].next
                self.n[pre].next=Null                    
                self.delete_index(pre)                     ##???? ptr이 들어가야 하는지 pre가 들어가야 하는지 헷갈립니다.
                self.current=pre                   
                self.no-=1
                
    def remove(self,p:int)->None:
        if self.head !=Null:
            if p== self.head:                           ##하나밖에 없을때
                self.remove_first()
            else:
                ptr=self.head
                
                while self.n[ptr].next != p:
                    ptr=self.n[ptr].next
                    if ptr==Null:                     ##못찾았을때
                        return
                self.delete_index(p)
                self.n[ptr].next=self.n[p].next
                self.currnet=ptr
                self.no-=1
                
    def remove_current_node(self)->None:
        self.remove(self.current)
        
    def clear(self)->None:
        while self.head !=Null:              ##첫번째부터 계속 지우는 과정
            self.remove_first()
        self.current=Null
    
    def next(self)->bool:
        if self.current==Null or self.n[self.current].next==Null: ##지금 노드가 비어있거나 다음이 비어있을때
            return False
        self.current=self.n[self.current].next
        return True
    
    def print_current_node(self)->None:
        if self.current==Null:
            print("주목 노드가 없다.")
        else:
            print(self.n[self.current].data)
            
    def print(self)->None:
        ptr=self.head
        while ptr !=Null:
            print(self.n[ptr].data)
            ptr=self.n[ptr].next
    
    def dump(self)->None:
        for i in self.n:
            print(f'[{i}] {i.data} {i.next} {i.dnext}')
    
    def __iter__(self)->ArrayLinkedListIterator:
        return ArrayLinkedListIterator(self.n, self.head)
    
class ArrayLinkedListIterator:
    def __init__(self, n:int , head: int):
        self.n=n
        self.current=head
    
    def __iter__(self)->ArrayLinkedListIterator:
        return self
    
    def __next__(self)->Any:
        if self.current==Null:
            raise StopIteration
        else:
            data=self.n[self.current].data
            self.current=self.n[self.current].next
            return data
        
        
```
