```python
from __future__ import annotations
from typing import Any,Type

class Node:
    def __init__(self,key:Any, value:Any,left:Node=None,right:Node=None):
        self.key=key
        self.value=value
        self.left=left   ##왼쪽자식
        self.right=right ##오른쪽 자식
        
class BinarySearchTree:
    
    def __init__(self):
        self.root=None
        
    def search(self,key:Any)->Any:
        p=self.root  ##루트
        while True:
            if p is None:              ##루트가없으면 진행불가
                return None
            if key==p.key:
                return p.value
            elif key<p.key:                  ##부모노드보다 key값이 작으면
                p=p.left
            else:                      ##부모노드보다 key값이 크면
                p=p.right
    
    def add(self,key:Any, value:Any)->bool:
        
        def add_node(node:Node,key:Any,value: Any)->None:
            if key==node.key:  
                return False        ##이미 존재
            elif key<node.key:          
                if node.left is None: ##왼쪽자식노드없으면 추가
                    node.left=Node(key,value,None,None)
                else:
                    add_node(node.left,key,value)
            else:
                if node.right is None: ##오른쪽자식노드 없으면 추가
                    node.right=Node(key,value,None,None)
                else:
                    add_node(node.right,key,value)
            return True
        
        if self.root is None: ##루트조차 없을때 
            self.root=Node(key,value,None,None)  
            return True
        else:
            return add_node(self.root,key,value)
        
    def remove(self,key:Any)->bool:
        p=self.root
        parent=None
        is_left_child=True
        
        while True:
            if p is None:           #루트가 없으면
                return False
            if key==p.key:
                break
            else:
                parent=p                 
                if key<p.key:           ##부모값보다 key가 작으면 왼쪽에
                    is_left_child=True
                    p=p.left
                else:                    ##부모값보다 key값이 크면 오른쪽에
                    is_left_child=False
                    p=p.right
                    
        if p.left is None:     ##왼쪽 자식이 없으면
            if p is self.root:
                self.root=p.right
            elif is_left_child:
                parent.left=p.right
            else:
                parent.right=p.right
                
        elif p.right is None:      ##오른쪽 자식이 없으면
            if p is self.root:
                self.root=p.left
            elif is_left_child:
                parent.left=p.left
            else:
                parent.right=p.left
        else:
            parent=p
            left=p.left                    ##가장 작은 노드설정
            is_left_child=True
            while left.right is not None: 
                parent=left
                left=left.right
                is_left_child=False
            
            p.key=left.key
            p.value=left.value
            if is_left_child:
                parent.left=left.left
            else:
                parent.right=left.left
        return True
    
    def dump(self,reverse=False)->None:
        
        def print_subtree(node:Node):
            if node is not None:
                print_subtree(node.left)        ##왼쪽서브트리 오름차순출력
                print(f'{node.key} {node.value}')       #노드 출력
                print_subtree(node.right)      ##오른쪽서브트리 오름차순출력
        
        def print_subtree_rev(node:Node): ##내림차순출력
            if node is not None:
                print_subtree_rev(node.right)
                print(f'{node.key} {node.value}')
                print_subtree_rev(node.left)
                
        print_subtree_rev(self.root) if reverse else print_subtree(self.root)
        
    
    def min_key(self)->Any:      ##가장 왼쪽 아래 자식 찾는과정
        if self.root is None: 
            return None
        p=self.root
        while p.left is not None:
            p=p.left
        return p.key
     
    def max_key(self)->Any:        ##맨 오른쪽 아래 자식 찾는 과정
        if self.root is None:
            return None
        p=self.root
        while p.right is not None:
            p=p.right
        return p.key
                
```
